```python
import sys #terminal interact
import pandas as pd
import time #discover how much time doesent come packages
import subprocess #call shell cripts subprocess.call([./shell.sh])
# print(subprocess.call(['./catch.sh']))
dump_file = './logs/brute_streams.csv'
global_count = 0;
list_time_tcp = []
list_time_udp = []
while True:
#     print(global_count)
#Reading stream file of tshark streams
    with open(dump_file, 'r') as file: #open file on read mode
        for aux in range(0,global_count):file.readline() #Jump already read lines
        lines = file.readlines(); #takes all lines of dump_file and send into $lines
    lines = [aux.strip() for aux in lines] #now line is a list where every position is a line and subposition is (not) an pkg feature

#Add pkgs to the corresponding stream file
    for line in range(global_count,len(lines)):
        if(lines[line][0] != ','): #if tcp, creates a tcp stream file with the stream number as name
            with open('./logs/tcp_csvs/tcp_stream_'+str(lines[line][0])+'.csv', 'a') as stream_file:
                stream_file.write(str(lines[line]+'\n'));
            try:list_time_tcp[lines[line][0]] = time.time() #each position on list_time is the correspondent number of stream, to indeitfy how much time that stream not receive conections
            except:list_time_tcp.append(time.time()) #if not exists, append it (probably will be the correspondent position on stream number)
            
        elif(lines[line][1] != ','):
            with open('./logs/udp_csvs/udp_stream_'+str(lines[line][1])+'.csv', 'a') as stream_file:
                stream_file.write(str(lines[line]+'\n'));
            try:list_time_udp[lines[line][0]] = time.time() #each position on list_time is the correspondent number of stream, to indeitfy how much time that stream not receive conections
            except:list_time_udp.append(time.time()) #if not exists, append it (probably will be the correspondent position on stream number)
    global_count += len(lines) #store's the number os lines already readed

#Verify the time that the stream doesent receive pkgs
    for count in range(0,len(list_time_tcp)):
        if((int(time.time()) - int(list_time_tcp[count])) >= 10):
            subprocess.call(['python3 ./handler.py ./logs/tcp_csvs/tcp_stream_'+count+'.csv tcp'])
    for count in range(0,len(list_time_tcp)):
        if((int(time.time()) - int(list_time_udp[count])) >= 10):
            subprocess.call(['python3 ./handler.py ./logs/udp_csvs/udp_stream_'+count+'.csv udp'])
```
